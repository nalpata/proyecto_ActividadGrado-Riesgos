## 3. Validaciones básicas

Se valida que no existan preguntas duplicadas y que los identificadores sean únicos.

```python
print('Total preguntas:', len(gold_questions))
print('IDs únicos:', gold_questions['question_id'].nunique())
print('Preguntas únicas:', gold_questions['question'].nunique())

if gold_questions['question_id'].duplicated().any():
    raise ValueError('Hay question_id duplicados')

if gold_questions['question'].duplicated().any():
    raise ValueError('Hay preguntas duplicadas')

print('Validación finalizada correctamente')
```

## 4. Resumen por categoría y tipo de tarea

```python
resumen_categoria = (
    gold_questions.groupby(['categoria', 'tipo_tarea'])
    .size()
    .reset_index(name='cantidad')
    .sort_values(['categoria', 'tipo_tarea'])
)

display(resumen_categoria)

resumen_dificultad = (
    gold_questions.groupby('nivel_dificultad')
    .size()
    .reset_index(name='cantidad')
)

display(resumen_dificultad)
```
