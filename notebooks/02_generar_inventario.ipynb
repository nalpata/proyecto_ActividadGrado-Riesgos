```python
# =====================================================
# Actividad de Grado MIA UC
# Patricia Patiño
# =====================================================

# Instalar dependencias
!pip -q install pandas openpyxl python-docx pymupdf

# Librerías
import pandas as pd
from pathlib import Path
from google.colab import files

print("======================================")
print("GENERADOR DE INVENTARIO DEL CORPUS")
print("======================================")

# -----------------------------------------------------
# Cargar documentos
# -----------------------------------------------------

print("\nSeleccione los documentos del corpus (PDF y DOCX)...")

uploaded = files.upload()

# -----------------------------------------------------
# Función clasificación documental
# -----------------------------------------------------

def clasificar_documento(nombre):

    nombre = nombre.lower()

    if "acta" in nombre:
        return "ACTA"

    elif "comunicado" in nombre:
        return "COMUNICADO"

    elif "incumplimiento" in nombre:
        return "INCUMPLIMIENTO"

    elif "mejora" in nombre:
        return "MEJORA"

    elif "review" in nombre:
        return "REVIEW"

    elif "ans" in nombre:
        return "ANS"

    else:
        return "OTRO"

# -----------------------------------------------------
# Construir inventario
# -----------------------------------------------------

registros = []

for i, archivo in enumerate(uploaded.keys(), start=1):

    ruta = Path(archivo)

    registros.append({
        "doc_id": f"DOC_{i:03d}",
        "filename": ruta.name,
        "extension": ruta.suffix.lower(),
        "tipo_documento": clasificar_documento(ruta.name)
    })

inventario = pd.DataFrame(registros)

# -----------------------------------------------------
# Mostrar inventario
# -----------------------------------------------------

print("\nPrimeros registros:\n")

display(inventario.head())

# -----------------------------------------------------
# Estadísticas
# -----------------------------------------------------

print("\n======================================")
print("ESTADÍSTICAS DEL CORPUS")
print("======================================")

print(f"\nTotal documentos: {len(inventario)}")

display(
    inventario["tipo_documento"]
    .value_counts()
    .reset_index()
    .rename(columns={
        "index":"tipo_documento",
        "tipo_documento":"cantidad"
    })
)

# -----------------------------------------------------
# Guardar CSV
# -----------------------------------------------------

inventario.to_csv(
    "inventario_corpus.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\nArchivo generado:")
print("inventario_corpus.csv")

# -----------------------------------------------------
# Descargar archivo
# -----------------------------------------------------

files.download("inventario_corpus.csv")

print("\nProceso finalizado.")
```
